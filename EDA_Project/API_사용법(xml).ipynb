```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
key = "76744d55566c686d36315647704d79"
url = f"http://openapi.seoul.go.kr:8088/{key}/xml/disabledCalltaxi/1/5/20190101"
```

```python
res=requests.get(url)
root=ET.fromstring(res.text)
```

```python
list1 = []
for i in root.iter("item"):
    dict1 = {}
    for j in i:
        dict1[j.tag] = j.text
    list1.append(dict1)
print(len(list1), list1)
```
